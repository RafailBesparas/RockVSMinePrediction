```python
import numpy as np
import pandas as pd

# Load the original dataset
original_dataset = pd.read_csv("Copy of sonar data.csv",header=None)

# Separate features and labels
features = original_dataset.iloc[:, :-1]
labels = original_dataset.iloc[:, -1]

# Introduce Gaussian noise to create synthetic data
noise = np.random.normal(0, 0.05, features.shape)  # Adjust noise level if necessary
synthetic_features = features + noise

# Combine synthetic features with original labels to form a new dataset
synthetic_dataset = pd.concat([synthetic_features, labels], axis=1)

# Save dataset to a csv
synthetic_dataset.to_csv('SyntheticData1.csv', index=False)  

# Save synthetic dataset to an xls file 
#data_to_excel = pd.ExcelWriter('SyntheticData1.xlsx')
#synthetic_dataset.to_excel(data_to_excel)
#data_to_excel.close()

synthetic_dataset.head()  # Display first few rows of the synthetic dataset
```
